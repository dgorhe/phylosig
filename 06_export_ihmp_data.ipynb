# Get rich T2D metadata

```python
t2d_data, t2d_metadata = get_diffs(
    "t2d", top_n=None, get_abundances=True, include_metadata=True
)
```

```python
t2d_subject_metadata = pd.read_table("/home/phil/DATA/ihmp/T2D/t2d_subject_metadata.tsv")

# Strip whitespace from column names
t2d_subject_metadata.columns = t2d_subject_metadata.columns.str.strip()

# Strip whitespace from values
for col in t2d_subject_metadata.columns:
    if t2d_subject_metadata[col].dtype == "object":
        t2d_subject_metadata[col] = t2d_subject_metadata[col].str.strip()

t2d_metadata_with_subjects = pd.merge(
    t2d_metadata,
    t2d_subject_metadata,
    left_on="subject_id",
    right_on="SubjectID"
)

t2d_metadata_with_subjects.to_csv("ihmp/t2d_metadata.csv.gz", compression="gzip")
```
