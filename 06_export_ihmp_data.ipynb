```python
# Site is 'nasal cavity' or 'feces' - first level of multiindex

t2d_ihmp_metadata = metadata
# Rename 'sample' column to 'filename'
t2d_ihmp_metadata = t2d_ihmp_metadata.rename(columns={"sample": "filename"})
t2d_ihmp_metadata = t2d_ihmp_metadata.reset_index()

t2d_ihmp_stool = t2d_ihmp_metadata[t2d_ihmp_metadata["site"] == "feces"]
t2d_ihmp_nasal = t2d_ihmp_metadata[t2d_ihmp_metadata["site"] == "nasal cavity"]

print(len(t2d_ihmp_stool), "stool samples")
print(len(t2d_ihmp_nasal), "nasal samples")
```

```python
# Great, now we need to recocile visit IDs with filenames

stool_ids = [(row["patient"], row["visit"]) for i, row in t2d_ihmp_stool.iterrows()]
nasal_ids = [(row["patient"], row["visit"]) for i, row in t2d_ihmp_nasal.iterrows()]

stool_ids = set(stool_ids)
nasal_ids = set(nasal_ids)

print(len(stool_ids), "stool IDs")
print(len(nasal_ids), "nasal IDs")
print(len(stool_ids.union(nasal_ids)), "total IDs")
```
