# Export iHMP datasets
> Just take the get_diffs function and output to directory

```python
from src.ihmp import get_diffs
```

```python
for name in ["ibd", "moms", "t2d"]:
    data, metadata = get_diffs(
        name, top_n=None, get_abundances=True, include_metadata=True
    )
    metadata = metadata.set_index(
        ["sample_body_site", "subject_id", "visit_number"]
    ).drop_duplicates()
    data.to_csv(f"ihmp/{name}_data.csv.gz", compression="gzip")
    metadata.to_csv(f"ihmp/{name}_metadata.csv")
```
