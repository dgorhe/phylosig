```python
# To read data:
import pandas as pd
import biom
import pyreadr
from Bio import SeqIO

import numpy as np
```

```python
otus_3c = pd.read_table(
    "diabimmune/diabimmune_3country_16s_otu_table.txt", index_col=0
)

otus_t1d = biom.load_table("diabimmune/diabimmune_t1d_16s_otu_table.biom")

otus_abx = pd.read_table(
    "diabimmune/diabimmune_antibiotics_otus.txt", index_col=0
)
```

```python
# Sanity check #1: OTU tables are additive

am_rows = [
    x
    for x in otus_abx.index
    if x.startswith(
        "k__Bacteria|p__Verrucomicrobia|c__Verrucomicrobiae|o__Verrucomicrobiales|f__Verrucomicrobiaceae|g__Akkermansia|s__muciniphila"
    )
]

am_sums = otus_abx.loc[am_rows, :].sum(axis=1)
print("species sum", am_sums.iloc[0])
print("sum of OTUs", am_sums.iloc[1:].sum())

# This looks like it's a tiny rounding error, so we can go ahead and ignore it
```

```python
# Sanity check #2: do this for all samples

for id in otus_abx.index:
    am_rows = [x for x in otus_abx.index if x.startswith(id)]

    am_sums = otus_abx.loc[am_rows, :].sum(axis=1)

    if len(am_sums) > 1:
        # print(f"{am_sums.iloc[0]:.3f}\t{am_sums.iloc[1:].sum():.3f}")
        if not np.isclose(am_sums.iloc[0], am_sums.iloc[1:].sum(), atol=1e-4):
            print(f"Error for {id}")
            print(f"{am_sums.iloc[0]:.3f}\t{am_sums.iloc[1:].sum():.3f}")
```

```python
gg_ids = [x.id for x in SeqIO.parse("greengenes/data/gg_13_5.fasta", "fasta")]

def clean_otu_table(otu_table):
    """Remove OTUs without greengenes IDs and trim"""
    otu_table.index = [x.split("|")[-1] for x in otu_table.index]
    intersect = set(otu_table.index) & set(gg_ids)
    otu_table = otu_table.loc[intersect, :]
    return otu_table

# Clean all
otus_3c_cleaned = clean_otu_table(otus_3c)
otus_abx_cleaned = clean_otu_table(otus_abx)
otus_t1d_cleaned = otus_t1d.to_dataframe()

# Normalize
otus_3c_cleaned = otus_3c_cleaned / otus_3c_cleaned.sum(axis=0)
otus_abx_cleaned = otus_abx_cleaned / otus_abx_cleaned.sum(axis=0)
otus_t1d_cleaned = otus_t1d_cleaned / otus_t1d_cleaned.sum(axis=0)
```

```python
otus_merged = pd.concat([otus_3c_cleaned, otus_abx_cleaned, otus_t1d_cleaned], axis=1).fillna(0)

otus_merged.to_csv("diabimmune/diabimmune_cleaned.csv")
```
